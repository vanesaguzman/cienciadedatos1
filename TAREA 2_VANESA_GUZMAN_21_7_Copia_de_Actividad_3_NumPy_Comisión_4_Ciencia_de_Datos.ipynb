# **Cules son las ventajas y desventajas de NumPy?**

Algunas de sus ventajas son:
- Tiene funciones matemticas y estadsticas integradas que facilitan el anlisis de datos.
- Permite realizar operaciones vectorizadas que evitan el uso de bucles y mejoran el rendimiento.
- Es compatible con otras bibliotecas de Python como pandas, scipy y matplotlib que amplan sus funcionalidades.

Algunas de sus desventajas son:
- Requiere aprender una sintaxis y una lgica diferentes a las de Python nativo.
- Puede consumir mucha memoria si se crean matrices muy grandes o se realizan operaciones complejas.
- No es tan flexible como Python para trabajar con tipos de datos heterogneos o estructuras anidadas.


# **Para qu usamos NumPy?**
Numpy se usa para automatizar tareas en un trabajo que usa grandes cantidades de datos porque permite procesarlos de forma rpida y precisa, reduciendo el tiempo y el esfuerzo necesarios.

Adems, facilita la visualizacin y la comunicacin de los resultados mediante grficos y tablas.

Numpy es una herramienta muy til para cualquier profesional que trabaje con datos numricos, como cientficos, ingenieros, analistas o programadores.

# **Cmo organizamos los datos para facilitar su visualizacin?**

Para organizar los datos para visualizarlos despus, usando numpy, podemos seguir estos pasos:

- Importamos la biblioteca numpy y otras que necesitamos para leer o escribir archivos de datos, como csv, json, etc.




```python
import numpy as np
```

- Creamos un arreglo numpy a partir de una lista de Python o de un archivo de datos usando funciones como np.array.


```python
#Antes de usar Numpy en un conjunto de datos,
#debemos convertirlo en un vector o arreglo.
#Array en NumPy es similar a la lista en Python.

#Est representado por llaves con valores
#en su interior separados por comas.

lista = [5, 10.2, 4, .3,6.6]

print(lista)
```

```python
#Luego podemos transformar el conjunto de datos
#en una arreglo de NumPy usando:

arreglo = np.array(lista)

print(arreglo)
```


- Aplicamos operaciones sobre el arreglo numpy para ordenar, filtrar, agrupar o transformar los datos segn las necesidades.



**Media**

Calculamos el promedio de los datos:

```python
promedio = np.mean(arreglo)

print(promedio)
```

```python
media = round(promedio)

print(media)

#Tambin podemos decir cuntos decimales queremos obtener
media = round(promedio, 3)

print(media)
```

```python
#Podemos usar la funcin NumPy np.sort para ordenar nuestros datos.

```

```python
#Podemos filtrar los datos

segmento_inicio = arreglo[0:3]

print(segmento_inicio)
```

```python
#Podemos calcular la mediana de los datos

mediana = np.median(arreglo)

print(mediana)
```

# **Cmo hacemos el grfico de los datos en un arreglo?**

```python
import numpy as np
import matplotlib.pyplot as plt

# Creamos el array de 5 nmeros
y = np.array([1, 3, 2, 4, 5])

# Creamos el array de 5 valores para el eje x
x = np.arange(1, 6)

# Grafica el array de nmeros
plt.plot(x, y)

# Mostramos el grfico
plt.show()
```
