Usar una lista de valores en Python tiene varias ventajas sobre usar muchas variables, entre ellas:

- Permite almacenar múltiples valores en una sola variable, lo que facilita el manejo y la organización de los datos.

- Consumen más memoria que las variables simples, ya que almacenan más información.

- Pueden contener elementos duplicados, lo que puede generar problemas de consistencia o redundancia de los datos.

- Pueden ser más lentas que otros tipos de datos como los arrays o los sets, ya que tienen más operaciones y verificaciones internas.


# **Por qué usamos la estructura de control condicional**
[Más info aquí](https://tutorialpython.com/estructuras-de-control-en-python/)

```python
# Defino una variable y un número
dinero = 100
minimo = 50

# Uso un if para comparar la variable con el número
if dinero >= minimo:
  # Devuelvo "me gusta"
  # si la variable es mayor o igual al dinero que quiero ganar.
  print("Me gusta")
else:
  # Devuelvo "no me gusta" si es menor a lo que quiero ganar
  print("No me gusta")
```

```python
# Código en limpio

dinero = 100
minimo = 50

if dinero >= minimo:
  print("Me gusta")
else:
  print("No me gusta")
```

Uso if en Python para tomar decisiones en mi programa según el valor de una o más condiciones. El if me permite ejecutar un bloque de código si la condición se cumple (es verdadera) y otro bloque de código si no se cumple (es falsa).

El if en Python es muy útil para controlar el flujo de mi programa y adaptarlo a diferentes situaciones o entradas. Por ejemplo, puedo usar un if para verificar si el usuario ha ingresado un dato válido, si un número es par o impar.

# **Por qué usamos funciones en Python**

Algunas de las ventajas de usar funciones en Python son:

- **Reducir la duplicación de código:** Las funciones permiten escribir el código una sola vez y usarlo muchas veces, evitando repetir el mismo código en diferentes partes del programa.

- **Mejorar la reutilización del código:** Las funciones se pueden llamar desde cualquier parte del programa, e incluso desde otras funciones, lo que facilita el uso del código existente para diferentes propósitos.

- **Escribir código más claro:** Las funciones ayudan a organizar el código en bloques lógicos y con nombres descriptivos, lo que hace que el código sea más fácil de leer y entender.

- **Descomponer problemas complejos en partes más simples:** Las funciones permiten dividir un problema grande en subproblemas más pequeños y manejables, lo que facilita el diseño, la depuración y el mantenimiento del código.

# **Qué es el bucle while de Python**

[Más info](https://tutorialpython.com/bucles-while-y-for-en-python/)

python-ciclo-while.jpg

El bucle while en Python es una estructura de control que se usa para ejecutar un bloque de código repetidamente mientras se cumple una condición. La sintaxis del bucle while es la siguiente:

    while condición:
        cuerpo del bucle

     

La condición se evalúa en contexto booleano, es decir, si es verdadera o falsa. Si la condición es verdadera, el cuerpo del bucle se ejecuta. Luego, la condición se vuelve a evaluar, y si sigue siendo verdadera, el cuerpo se ejecuta de nuevo. Esto continúa hasta que la condición sea falsa, en cuyo caso la ejecución del programa continúa con la primera línea después del bucle.



```python

# Defino las variables
"""
dinero1 = 5
dinero2 = 10
dinero3 = 7
dinero4 = 100
dinero5 = 20
"""
"""
# Creo una lista con las variables
lista1 = [dinero1, dinero2, dinero3, dinero4, dinero5]

# Creo un contador para recorrer la lista
contador = 0

# Creo un bucle while que se ejecute mientras el contador sea menor que 3
# Solamente quiero ver los primeros 3 trabajos

while contador < 3:
  # Muestro el valor de la variable en la posición del contador
  print(lista1[contador])
  # Incremento el contador en 1
  contador += 1
  """

  num = 10
  while num <20
  print (num)
  num += 1
```

```python
#Código en limpio

lista1 = [dinero1, dinero2, dinero3, dinero4, dinero5]
contador = 0

while contador < 3:
  print(lista1[contador])
  contador += 1
else:
  print('Fin')
```

Dentro del bucle while, se pueden usar las sentencias break y continue para alterar el flujo normal del bucle. La sentencia break sirve para terminar el bucle prematuramente, mientras que la sentencia continue sirve para saltar a la siguiente iteración sin ejecutar el resto del cuerpo.

# **Qué es el ciclo For de Python**

[Más info](https://tutorialpython.com/bucles-while-y-for-en-python/)

python-ciclo-for.jpg

El ciclo for de Python es una estructura de control que se utiliza para iterar sobre un objeto iterable y ejecutar un bloque de código para cada elemento en la secuencia de dicho objeto. La sintaxis básica del ciclo for en Python es la siguiente:

**for elemento in iterable:**

   **código a ejecutar por cada elemento**

Donde:

- Elemento es la variable que toma el valor de cada elemento en la secuencia.

- Iterable es un objeto iterable, como una lista, una tupla, un diccionario, un conjunto o una cadena.

El ciclo for permite recorrer cualquier objeto iterable de manera sencilla y eficiente, y realizar operaciones repetitivas con cada elemento.

```python
# Cuando uso el ciclo for
# no necesito una variable contador
# Quiero mostrar mi lista completa

"""
dinero1 = 5
dinero2 = 10
dinero3 = 7
dinero4 = 100
dinero5 = 20
"""

lista1 = [dinero1, dinero2, dinero3, dinero4, dinero5]

for elemento in lista1:
  print(elemento)

# Por cada elemento en mi lista
# mostrar ese elemento
```

```python
# Quiero mostrar solamente los primeros 3 elementos de mi lista
# También puedo usar la función range(0,3)

"""
dinero1 = 5
dinero2 = 10
dinero3 = 7
dinero4 = 100
dinero5 = 20
"""

lista1 = [dinero1, dinero2, dinero3, dinero4, dinero5]
contador = 0

for elemento in range(0,3):
  print(lista1[elemento])
```

# **Cómo hago una función de Python**

Se puede hacer lo siguiente:

- Defino el nombre de la función y los parámetros que recibe.

En este caso, la función podría llamarse sumar_lista y recibir una lista como parámetro.

- Indentar el código que se quiere ejecutar dentro de la función debajo de la declaración de la función.

- Usar el parámetro de la función en lugar de la variable lista1.

- Usar la palabra clave return para devolver el resultado de la suma en lugar de imprimirlo.

```python
# Defino mi función

def sumar_lista(lista):
  contador = 0
  for elemento in lista:
    contador += elemento
  return contador
```

```python
# Llamada a la función

lista1 = [dinero1, dinero2, dinero3, dinero4, dinero5]
resultado = sumar_lista(lista1)
print(f"Sumando todos los trabajos, gano {resultado} dólares en total.")
```
