# **Qu es el bucle while de Python**

[Ms info](https://tutorialpython.com/bucles-while-y-for-en-python/)

python-ciclo-while.jpg

El bucle while en Python es una estructura de control que se usa para ejecutar un bloque de cdigo repetidamente mientras se cumple una condicin. La sintaxis del bucle while es la siguiente:

    while condicin:
        cuerpo del bucle

     

La condicin se evala en contexto booleano, es decir, si es verdadera o falsa. Si la condicin es verdadera, el cuerpo del bucle se ejecuta. Luego, la condicin se vuelve a evaluar, y si sigue siendo verdadera, el cuerpo se ejecuta de nuevo. Esto contina hasta que la condicin sea falsa, en cuyo caso la ejecucin del programa contina con la primera lnea despus del bucle.



```python

# Defino las variables
"""
dinero1 = 5
dinero2 = 10
dinero3 = 7
dinero4 = 100
dinero5 = 20
"""
"""
# Creo una lista con las variables
lista1 = [dinero1, dinero2, dinero3, dinero4, dinero5]

# Creo un contador para recorrer la lista
contador = 0

# Creo un bucle while que se ejecute mientras el contador sea menor que 3
# Solamente quiero ver los primeros 3 trabajos

while contador < 3:
  # Muestro el valor de la variable en la posicin del contador
  print(lista1[contador])
  # Incremento el contador en 1
  contador += 1
  """

  num = 10
  while num <20
  print (num)
  num += 1
```

```python
#Cdigo en limpio

lista1 = [dinero1, dinero2, dinero3, dinero4, dinero5]
contador = 0

while contador < 3:
  print(lista1[contador])
  contador += 1
else:
  print('Fin')
```

Dentro del bucle while, se pueden usar las sentencias break y continue para alterar el flujo normal del bucle. La sentencia break sirve para terminar el bucle prematuramente, mientras que la sentencia continue sirve para saltar a la siguiente iteracin sin ejecutar el resto del cuerpo.

# **Qu es el ciclo For de Python**

[Ms info](https://tutorialpython.com/bucles-while-y-for-en-python/)

python-ciclo-for.jpg

El ciclo for de Python es una estructura de control que se utiliza para iterar sobre un objeto iterable y ejecutar un bloque de cdigo para cada elemento en la secuencia de dicho objeto. La sintaxis bsica del ciclo for en Python es la siguiente:

**for elemento in iterable:**

   **cdigo a ejecutar por cada elemento**

Donde:

- Elemento es la variable que toma el valor de cada elemento en la secuencia.

- Iterable es un objeto iterable, como una lista, una tupla, un diccionario, un conjunto o una cadena.

El ciclo for permite recorrer cualquier objeto iterable de manera sencilla y eficiente, y realizar operaciones repetitivas con cada elemento.

```python
# Cuando uso el ciclo for
# no necesito una variable contador
# Quiero mostrar mi lista completa

"""
dinero1 = 5
dinero2 = 10
dinero3 = 7
dinero4 = 100
dinero5 = 20
"""

lista1 = [dinero1, dinero2, dinero3, dinero4, dinero5]

for elemento in lista1:
  print(elemento)

# Por cada elemento en mi lista
# mostrar ese elemento
```

# **Cmo hago una funcin de Python**

Se puede hacer lo siguiente:

- Defino el nombre de la funcin y los parmetros que recibe.

En este caso, la funcin podra llamarse sumar_lista y recibir una lista como parmetro.

- Indentar el cdigo que se quiere ejecutar dentro de la funcin debajo de la declaracin de la funcin.

- Usar el parmetro de la funcin en lugar de la variable lista1.

- Usar la palabra clave return para devolver el resultado de la suma en lugar de imprimirlo.

```python
# Defino mi funcin

def sumar_lista(lista):
  contador = 0
  for elemento in lista:
    contador += elemento
  return contador
```

```python
# Llamada a la funcin

lista1 = [dinero1, dinero2, dinero3, dinero4, dinero5]
resultado = sumar_lista(lista1)
print(f"Sumando todos los trabajos, gano {resultado} dlares en total.")
```
